# Vektoren

```python
import numpy as np

#Erzeugung per Hand
x = np.array( [5,6,7,8] );

#Automatisch
x = np.ones(10); print(x)
x = np.zeros(10); print(x)

#Spannender
x = np.arange(1,2,0.1); print(x)
x = np.linspace(1,2,20); print(x)

# Noch spannender...
x = np.random.random(10); print(x)
x = np.random.normal(0,1,10); print(x)

```

#Beispiel

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.linspace(-6, 6, 40)

sns.set()
fig,ax = plt.subplots(figsize=(9, 9))

ax.set_title("Sinus") 
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
#ax.set_aspect("equal")

sns.lineplot(x=x, y= x*x*x-2*x, color="blue", label="sin(x)")
```
